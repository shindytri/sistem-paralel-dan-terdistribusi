```python
# CLIENT Upload

# import library socket karena akan menggunakan IPC socket
import socket

# definisikan tujuan IP server
TCP_IP = "192.168.43.40"

# definisikan port dari server yang akan terhubung
TCP_PORT = 11720


# definisikan ukuran buffer untuk mengirimkan pesan
BUFFER_SIZE = 1024

# definisikan pesan yang akan disampaikan
PESAN = "Hai Shindul"

# buat socket TCP
s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)

# lakukan koneksi ke server dengan parameter IP dan Port yang telah didefinisikan
s.connect((TCP_IP,TCP_PORT))

# kirim pesan ke server
s.send(PESAN.encode())

# terima pesan dari server
data = s.recv(BUFFER_SIZE)

# tampilkan pesan/reply dari server
print("Data dari server : ", data.decode())

# Upload File
f = open("file_diupload.txt", 'rb')
l = f.read(BUFFER_SIZE)  # each time we only send 1024 bytes of data
while (l):
    print('File uploading...')
    s.sendall(l)
    l = f.read(BUFFER_SIZE)

print("File Uploaded Successfully!")


# tutup koneksi
s.close()
```

```python
# SERVER

# import library socket karena akan menggunakan IPC socket
import socket

# definisikan alamat IP binding  yang akan digunakan 
TCP_IP = '192.168.43.201'

# definisikan port number binding  yang akan digunakan 
TCP_PORT = 5005

# definisikan ukuran buffer untuk mengirimkan pesan
BUFFER_SIZE = 2048

# buat socket bertipe TCP
s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)

# lakukan bind
s.bind((TCP_IP, TCP_PORT))

# server akan listen menunggu hingga ada koneksi dari client
s.listen(1)

# lakukan loop forever
while 1:
    # menerima koneksi
    conn, addr = s.accept()
    # menampilkan koneksi berupa IP dan port client yang terhubung menggunakan print
    print('Alamat : ', addr)
    # menerima data berdasarkan ukuran buffer
    data= conn.recv(BUFFER_SIZE)
    # menampilkan pesan yang diterima oleh server menggunakan print
    print('Data diterima : ', data.decode())
    # mengirim kembali data yang diterima dari client kepada client
    conn.send(data)

# tutup koneksi	
conn.close()
```
